```python
# Calculate the 75th and 25th percentile using the quantile function

cols = ["Age", "MonthlyIncome", "YearsAtCompany", "TrainingTimesLastYear"]
q3 = df[cols].quantile(0.75)
q1 = df[cols].quantile(0.25)

iqr = q3 - q1
lowerbound = q1 - (1.5*iqr)
upperbound = q3 + (1.5*iqr)

df[cols] = df[cols].clip(lower=lowerbound, upper=upperbound, axis=1)
df[cols]
```

### Oversampling Method for Resampling

```python
from imblearn.over_sampling import SMOTE
```

```python
# Split our samples to X and Y (axis) for the sampling
x = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

x = features we are using to predict the target variable

y = target variable (what we are trying to predict)

```python
# Resampling
smote = SMOTE(random_state=42)
X_resample, y_resample = smote.fit_resample(x,y)
```

```python
y_resample.value_counts()
```

```python
sns.countplot(x=y_resample);
```

### Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X_resample, y_resample, random_state=42, test_size=0.2, stratify=y_resample)
```

X_train (features) + y_train (prediction) = Train the Model -----> Exposed to the model

X_test + y_test = Test/Evaluate the dataset -----> Not exposed to the model

```python
print("Training set size", len(X_train))
print("Test set size", len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Build first model
model_LR = LogisticRegression(random_state=42)
model_LR.fit(X_train, y_train)
```

```python
# Train second model
model_RF = RandomForestClassifier(random_state=42)
model_RF.fit(X_train, y_train)
```

### Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_pred = model_LR.predict(X_train)

print(classification_report(y_pred, y_train))
```

```python
cm = confusion_matrix(y_pred, y_train)

sns.heatmap(cm, annot=True);
```

```python
y_pred = model_LR.predict(X_test)

print(classification_report(y_pred, y_test))
```

```python
cm = confusion_matrix(y_pred, y_test)

sns.heatmap(cm, annot=True, fmt="g");
```

```python
y_pred = model_RF.predict(X_train)

print(classification_report(y_pred, y_train))
```

```python
cm = confusion_matrix(y_pred, y_train)

sns.heatmap(cm, annot=True, fmt="g");
```

```python
y_pred = model_RF.predict(X_test)

print(classification_report(y_pred, y_test))

cm = confusion_matrix(y_pred, y_test)

sns.heatmap(cm, annot=True, fmt="g");
```
